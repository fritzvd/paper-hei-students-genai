# Perceived Privacy, Anonimyty and Disinhibition in HEI students

Fritz Van Deventer (HAN University of Applied Sciences)

## Introduction

Recent developments have seen generative Artificial Intelligence (genAI) being available for the masses with Large Language Models (LLM’s) having a much easier interface for non-technical users. Often signified with the moniker: AI or with magical stars (✨) it features in many applications ranging from web search to ([Google n.d.](#ref-googleGoogleAIMode); [DuckDuckGo n.d.](#ref-duckduckgoDuckaiDuckDuckGoHelp); [Microsoft n.d.](#ref-microsoftIntroducingNewBing)) personal communication such as Outlook ([“Chat with Copilot in Outlook - Microsoft Support” n.d.](#ref-ChatCopilotOutlook)) or Whatsapp ([Meta n.d.](#ref-metaMetaAIWhatsApp)). genAI is quickly becoming embedded in all sorts of business and applications ([Grantham-Philips 2024](#ref-grantham-philipsAIBecomingIngrained2024)). Even more than in applications beforehand users are encouraged to upload documents and provide more context for the LLM to perform better.

As a result of the increased use of genAI, such as LLM’s, a great variety of topics have been researched in the last few years. Such as environmental impact ([Ding and Shi 2024](#ref-dingSustainableLLMServing2024); [Rillig et al. 2023](#ref-rillig2023risks)), closing and / or widening of the digital divide ([Carter, Liu, and Cantrell 2020](#ref-carterExploringIntersectionDigital2020); [Sartori and Bocca 2023](#ref-sartoriMindingGapsPublic2023)). In the context of education, and higher education specifically research has been conducted concerning a number of topics including: guidelines and policy ([Wang et al. 2024](#ref-wangGenerativeAIHigher2024)), about usage, knowledge and skills ([Suonpää, Heikkilä, and Dimkar 2024](#ref-suonpaaStudentsPerceptionsGenerative2024); [Silvennoinen, Aksovaara, and Alanko-Turunen 2025](#ref-silvennoinenStudentsUsageGenAI2025)), personalized learning ([Zhang, Yao, and Hadizadeh Moghaddam 2025](#ref-zhangDesigningGenAITools2025); [Udeh 2025](#ref-udehRoleGenerativeAI2025)), the effect on learning outcomes ([Han, Peng, and Liu 2025](#ref-hanImpactGenAILearning2025)) and the use of genAI in collaborative learning ([Perifanou and Economides 2025](#ref-perifanouCollaborativeUsesGenAI2025)).
